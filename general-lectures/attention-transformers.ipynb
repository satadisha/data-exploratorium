### Limits to Recurrent Neural Networks

<div style="display:flex; align-items:flex-start; gap:20px;">
  <div class="fragment" style="flex:1; font-size:14px;">
    <p>Pre-transformer encoder and decoder models were often both Recurrent Neural Networks (RNNs).
    <ul>
      <li>They process one token at a time, sequentially.</li>
    </ul>
    </p>
    <p>An RNN takes two vectors at each time step:
    <ul>
      <li>An input vector (for the encoder, one word from the input sentence)</li>
      <li>A hidden state
        <ul>
          <li>It incorporates information from previous words together with the current one</li>
          <li>At each time step, the RNN updates its hidden state based on the current input and previous hidden state</li>
        </ul>
      </li>
    </ul>
    </p>
    <p>Model choices made at setup time include:
    <ul>
      <li>Size of the context vector, often 256, 512, or 1024</li>
      <li>Number of hidden units in the encoder RNN</li>
    </ul>
    </p>

  </div>

### Limitations of RNNs:

- Difficulty Capturing Long-Range Dependencies
    - Vanishing gradients.
    - Weak long-distance connections.
    - Difficulty establishing long-distance relationships.
        - `The animal didn't cross the street because it was too tired` 
        - <i>What does “it” in this sentence refer to? Is it referring to the street or to the animal?</i>
- Encoder Bottleneck
    - All information about the sentence must be stored in one vector.
- Sequential Processing
    - RNNs process tokens one at a time which can be slow for long sequences.

- Rather than compressing the context into a single vector, decoder should want to revisit the input sequence at every step. 
- At particular decoding steps, decoder should <u style="color: blue;">selectively attend</u> certain parts of the input sequence.
    - Simulate this as consuming a weighted sum of the representations of the input at each time step
    - Weights determine the extent to which each step’s context “focuses” on each input token
    - Can the weight assignment process be a differentiable function?
        - Can be learned along with all of the other neural network parameters.

### Going away from RNN Encoder

An attention model differs from a classic sequence-to-sequence model in **two** main ways.

**First**, the encoder passes a lot more data to the decoder 
- not just the last hidden state of the encoding stage
- the encoder passes all the hidden states to the decoder

### Going away from RNN Decoders

**Attention decoder** performs an additional step before producing its output.<br> 
Focus on the parts of the input most relevant for the current decoding step:
<div style="display:flex; gap:30px; align-items:flex-start;">

<div style="flex:1;">
<ul>
<li>Receives a set of <b>encoder hidden states</b>: each corresponds roughly to a token in the input sentence.</li>
<li>Assigns a <b>score</b> to each hidden state.</li>
<li>Applies <b>softmax</b> to convert these scores into attention weights.</li>
<li>Multiplies each hidden state by its weight, amplifying important states and suppressing less relevant ones.</li>
</ul>

</div>

<div style="display:flex; gap:1px; align-items:flex-start;">

<div style="flex:1;">

First the attention decoder RNN takes as input:

<ul>
<li>The encoder embedding of the <b>&lt;END&gt;</b> token</li>
<li>An initial decoder hidden state</li>
</ul>

At each time step of decoding, it produces: <br> 
1) an output <br> 
2) a new hidden state vector <b>$h_4$</b><br>

<em>The raw RNN output is discarded.</em>

<div style="display:flex; gap:30px; align-items:flex-start;">

<b>Attention Step</b>
<ul>
<li>Use the encoder hidden states and the decoder state $h_4$ to compute a context vector $C_4$.</li>
<li>Concatenate $h_4$ and $C_4$.</li>
<li>Pass this vector through a <b>Feedforward Neural Network (FFNN)</b>.</li>

### Why self-attention?
<div>
Cross-attention <b>connects two sequences.</b> 

Self-attention models <b>relationships within the same sequence.</b>
- Untethered to the RNN sequential architecture.<br>
</div>

<div style="display:flex; gap:20px; align-items:flex-start;">

<div style="width:50%;">

#### Cross-Attention

<div>Used when <b>one sequence attends to another sequence</b>.

Example: **Machine Translation**
</div>


**Encoder (input)**
The cat sits on the mat<br>

**Decoder (output)**
Le chat est assis<br>

When predicting **chat**, the decoder attends to the encoder tokens.

<br>

</div>

<div style="width:50%;">

#### Self-Attention

<div>

Used when **tokens attend to other tokens in the same sequence**.

</div>

<div>
Example sentence: <i>The <b>animal</b> didn't cross the street because <b>it</b> was tired.</i><br>

When computing the representation for **it**, the model compares it with:

- animal  
- street  
- tired  

<br>
</div>

### Self-Attention in Detail

<div style="display:flex; gap:30px; align-items:flex-start;">

<div style="flex:1;">

Start with your regular *off-the-shelf* word embeddings.

From each encoder input vector (each word embedding), $3$ **new vectors** (<i>projections</i>) are learned.

<ul>
<li>Produced by multiplying the embedding by three learned matrices:</li>
<ul>
<li>$\color{blue}{\text{Query}}$</li>
<li>$\color{blue}{\text{Key}}$</li>
<li>$\color{blue}{\text{Value}}$</li>
</ul>
</ul>

</div>

<div style="flex:1.5; text-align:center;">

<img src="media/images/self-attention-step1.png" width="500">

</div>

</div>

The dimensionality of these vectors often **smaller than embedding size**.
- Example: Query/Key/Value dimension = 64</li>
- Embedding dimension = 512
This makes the computation of **multi-head attention** more efficient.

### Self-Attention in Detail

<div style="display:flex; gap:30px; align-items:flex-start;">

<div style="flex:1;">

The **second step** is to calculate a score. Suppose we compute the self-attention for the first word **“Thinking”**.
<ul>
<li>Score each word in the input sentence against this word.</li>
<li>Determines how much the model should focus on other parts of the sentence when encoding this word.</li>
</ul>

The score is computed using the **dot product** between the query vector and key vectors. For the word at position $1$:
<ul>
<li>First score: $q_1 \cdot k_1$</li>
<li>Second score: $q_1 \cdot k_2$</li>
</ul>
</div>

<div style="flex:1; text-align:center;">

<img src="media/images/self-attention-full.png" width="420">

</div>

</div>

The **third step**:
- Divide scores by $\sqrt{d}$, where $d$ is the dimension of the key.
- This stabilizes the variance of the dot products.

### Self-Attention in Detail

<div style="display:flex; gap:30px; align-items:flex-start;">

<div style="flex:1;">

The **fourth step**:
<ul>
<li>Apply a <b>softmax</b> function.
<ul>
<li>Normalizes scores so they are positive and sum to $1$.</li>
<li>Determines how much each word should matter.</li>
</ul>
</li>
</ul>

The **fifth step**:
<ul>
<li>Multiply each value vector by its softmax score.</li>
<li>This amplifies important words and downweights less relevant ones.</li>
</ul>

The **sixth step**: Sum the weighted value vectors.

</div>

<div style="flex:1; text-align:center;">

<img src="media/images/self-attention-full.png" width="420">

</div>

</div>

### Matrix Calculation of Self-Attention

<div style="display:flex; gap:30px; align-items:flex-start;">

<div style="flex:1;">

To speed up computation, self-attention is implemented using **matrix operations**.

<ul>
<li>First, calculate the <b>Query</b>, <b>Key</b>, and <b>Value</b> matrices.</li>
<li>Pack the input embeddings into a matrix <b>$X$</b>.</li>
<li>Multiply $X$ by the learned weight matrices 

$W_Q$, $W_K$, and $W_V$</li>
</ul>

$$
Q = XW_Q, \qquad K = XW_K, \qquad V = XW_V
$$
</div>

<div style="flex:1; text-align:center;">

<img src="media/images/self-attention-calculations-1.png" width="420" style="margin-bottom:20px;">

</div>

</div>

### Matrix Calculation of Self-Attention


Now steps $2$ through $6$ can be written compactly as:

$$
\text{Attention}(Q,K,V)
=
\text{softmax}\left(
\frac{QK^T}{\sqrt{d_k}}
\right)V
$$
<img src="media/images/self-attention-calculations-2.png" width="420">

### Multi-Head Attention

<div style="display:flex; gap:30px; align-items:flex-start;">

<div style="flex:1;">

In practice, given the same set of queries, keys, and values, we may want the model to capture **different types of relationships** within a sequence.

For example, different attention mechanisms might focus on:
<ul>
<li><b>Short-range dependencies</b> (nearby words)</li>
<li><b>Long-range dependencies</b> (words far apart in the sentence)</li>
</ul>
</div>

<div style="flex:1.5; text-align:center;">
<img src="media/images/multiple-attention-heads.png" width="420">
</div>

</div>
The goal is to obtain different <b>representation subspaces</b> of the input embeddings.<br>
<b>Solution: Multi-head attention</b>

### Multi-Head Attention

<div style="display:flex; gap:30px; align-items:flex-start;">

<div style="flex:1;">

Instead of performing a <b>single attention operation</b>, the model performs multiple attention operations (called <em>heads</em>) simultaneously.

Each head:
<ul>
<li>Learns its own linear projections of <b>queries, keys, and values</b></li>
<li>Operates in a <b>different representation subspace</b></li>
</ul>

</div>

<div style="flex:1.5; text-align:center;">
<img src="media/images/multiple-attention-heads.png" width="420">
</div>

</div>

### Multi-Head Attention

1. **Project inputs into multiple subspaces**

For each attention head \(i\):
$$
Q_i = XW_Q^{(i)}, \quad
K_i = XW_K^{(i)}, \quad
V_i = XW_V^{(i)}
$$

2. **Compute attention independently for each head**
$$
\text{head}_i =
\text{Attention}(Q_i, K_i, V_i)
$$

3. **Concatenate the outputs of all heads**
$$
\text{Concat}(\text{head}_1, \dots, \text{head}_h)
$$

4. **Project back to the original representation space**

- Multiply by a learned weight matrix $W^O$.

### Positional Encoding

With RNNs, it was possible to preserve positional information. 
- Recurrently processes sequences one token at a time

Self-attention ditches this sequential approach for parallel computation.

What do we do if it really matters that the model knows in which order the input sequence arrived?
- Supplement token representations (embeddings) with additional vectors for positional information.
- Similar dimensionality as token embeddings but learns pattern to reveal the relative positions of tokens at a particular time step.
- Helps understand meaningful distances between the embedding vectors once they’re projected into $Q$/$K$/$V$ vectors for dot-product attention

### Transformers: Putting Everything Together

<div style="display:flex; gap:30px; align-items:flex-start;">

<div style="flex:1;">

**Rethinks the traditional encoder–decoder architecture** using self-attention. The architecture consists of two stacks:
<ul>
<li><b>Encoder stack</b> (6 layers)</li>
<li><b>Decoder stack</b> (6 layers)</li>
<li>Each token has its own pathway through the network.</li>
<li>This allows tokens to be processed <b>in parallel</b>, unlike RNNs which process tokens sequentially.</li>
</ul>

### Transformers: Encoder

<div style="display:flex; gap:30px; align-items:flex-start;">

<div style="flex:1;">

The **encoder layer** has two main components.<br>
**1. Self-Attention**
- Each word can attend to **other words in the sentence**
- Queries, keys, and values come from the **previous encoder layer**

**2. Feed-Forward Network**
- A small neural network applied **independently to each token**
- The same network is reused at every position (position-wise)

</div>

<div style="flex:1; text-align:center;">

<img src="media/images/transformer-encoder.png" width="420">

</div>

</div>

### Transformer: Encoder

Each component is wrapped with:<br>
**Residual Connection**
- Adds the input back to the output
- Helps preserve information and stabilize deep networks

**Layer Normalization**
- Normalizes features **within each token** (across dimensions)
- Keeps activations in a stable range during training
- LayerNorm works well when dealing with variable sequence lengths.


### Transformers: Decoder

Similar to the encoder but includes an additional attention step.

**1. Masked Self-Attention**
- Each token attends only to **previous tokens**
- Ensures the decoder generates text **left-to-right**

**2. Encoder–Decoder Attention**
- The decoder attends to the **encoder outputs**
- Encoder outputs provide **Keys (K)** and **Values (V)**
- Helps the decoder focus on the **relevant parts of the input sentence**

**3. Feed-Forward Network**
- A position-wise neural network applied **independently to each token**

As in the encoder:
- Inputs use **word embeddings + positional encoding**
- **Residual connections and layer normalization** stabilize training

### Masked Self-Attention

<div style="display:flex; gap:30px; align-items:flex-start;">

<div style="flex:1;">

In the **decoder**, the self-attention layer is only allowed to attend to **earlier positions** in the output sequence.<br>
Future positions are **masked** (set to `-inf`) before the softmax step in the attention calculation.<br>
**Why? → Decoding is an autoregressive task**
<ul>
<li>At inference time, the model must use only <b>past tokens</b> to predict the next token.</li>
<li>If full (bidirectional) attention were allowed, the model could indirectly access the token it is trying to predict.</li>
</ul>
</div>

<div style="flex:1; text-align:center;">
<img src="media/images/masked-attention.png" width="250">
</div>

</div>

<b>Key idea:</b>  
Masked attention prevents **information leakage from future tokens** during both training and inference.

### What would go wrong without masking?

If full (bidirectional) attention were allowed, the decoder could attend to **future tokens** during training.

Example: *The cat sat on the ___*

If the correct token is **"mat"**, the model could simply attend to that token and learn a trivial shortcut.

This would create <u style="color: blue;">information leakage</u>
- model would have access to information that wouldn't be available at inference time

### How masking works

Before the softmax in attention we compute
$$
\text{softmax}\left(\frac{QK^T}{\sqrt{d_k}} + M\right)
$$
where $M$ is a **causal mask**:
$$
M =
\begin{bmatrix}
0 & -\infty & -\infty & -\infty \\
0 & 0 & -\infty & -\infty \\
0 & 0 & 0 & -\infty \\
0 & 0 & 0 & 0
\end{bmatrix}
$$

This mask forces attention weights for future positions to become zero after the softmax.

Therefore, token $t$ can only attend to positions $\le t$.


### Transformers: Decoder

<div style="display:flex; gap:30px; align-items:flex-start;">

<div style="flex:1;">

Output of the decoder stack is a **vector of floats**.  
To convert this into a word prediction, the model uses:

**1. Linear Layer**
- Projects the decoder output into a large **logits vector**
- Size of the vector = **vocabulary size**

</div>

<div style="flex:1.5; text-align:center;">

<img src="media/images/transformer_decoder_output_layer.png" width="420">

</div>

</div>

**2. Softmax Layer**
- Converts logits into **probabilities**
- All probabilities are **positive** and **sum to 1**

The word with the **highest probability** becomes output token for that step.

### Training a Transformer

Learn model parameters that maximize the probability of the training sequences. The model predicts the **next token** given previous tokens. For an output sequence:
$$
P(y_1, y_2, \dots, y_T) = \prod_{t=1}^{T} P(y_t \mid y_1, \dots, y_{t-1})
$$
Training Procedure:
1. Start with a large dataset of text sequences
2. Feed tokens into the transformer
3. Predict the **next token** at each position
4. Compute **cross-entropy loss**: measures how different the model’s predicted probability distribution is from the true distribution
5. Update parameters using **gradient descent**

### Training a Transformer

Example:<br>
**Input**: The cat sat on the<br>
**Target**: cat sat on the mat<br>

The model predicts:
$$
P(cat | \text{The})\rightarrow P(sat | \text{The cat})\rightarrow P(on | \text{The cat sat})
$$

Decoding can be:
- Greedy: Only select token with highest probability at each step.
- Beam Search: hold on to, say, the top two words. 
    - say, ‘I’ and ‘a’ for example. 
    - Run the model twice: once assuming the first output, and another time assuming the second. Whichever version produced less error upon including the next step is kept.

### Pretrained Transformers and Large Language Models

Modern transformers are *pre*-trained on **massive text corpora**.

This process allows the model to learn **general language representations**, and
- grammar
- semantics
- factual knowledge
- reasoning patterns

After pretraining, the model can be adapted to new tasks using:
- **fine-tuning**
- **prompting**
- **instruction tuning**
