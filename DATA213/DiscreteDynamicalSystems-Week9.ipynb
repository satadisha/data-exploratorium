## Understanding population dynamics

<div style="display: flex; align-items: center; gap: 2px;">
  
  <div style="text-align: left; padding: 0;">
   <p>In 1990, mathematical ecologists intensified their drive to understand the population dynamics of the spotted owl.</p>
   
   <p>The life cycle of a spotted owl divides naturally into three stages:
    <ul>
    <li>Juvenile: up to 1 year old</li>
    <li>Sub-adult: 1 to 2 years</li>
    <li>Adult: older than 2 years</li>
    </ul>
   </p>

   <p>A first step in studying the population dynamics is to model the population at yearly intervals, at times denoted by $k=0, 1, 2, \dots$. Usually, 1:1 ratio of males to females assumed in each life stage and then count only the females.

### Understanding population dynamics

Let the population vector be 
$\mathbf{x}_k = (j_k, s_k, a_k)$, 
where $j_k$, $s_k$, and $a_k$ are the numbers of females in the juvenile, subadult, and adult stages, respectively.

Using actual field data from demographic studies, researchers build the following *stage-matrix model*:
$$\begin{pmatrix}
j_{k+1} \\
s_{k+1} \\
a_{k+1}
\end{pmatrix}
=
\begin{pmatrix}
0 & 0 & 0.33 \\
0.18 & 0 & 0 \\
0 & 0.71 & 0.94
\end{pmatrix}
\begin{pmatrix}
j_k \\
s_k \\
a_k
\end{pmatrix}$$

Here, $33\%$ of adult females in year $k$ (based on the average birth rate per owl pair) produces a new juvenile female in year $k+1$. Also,
- $18\%$ of the juveniles survive to become subadults,
- $71\%$ of the subadults survive to remain in the adult stage,
- $94\%$ of the adults survive to be counted as adults in the next year.

### Understanding population dynamics

The stage-matrix model is a difference equation of the form 
$\mathbf{x}_{k+1} = A \mathbf{x}_k$. This equation is often called a <b>discrete linear dynamical system</b>.
- It models the change in a system over the passage of finite but discrete amount of time.
- We want to understand long-term behaviors of such system. Here specifically to understand if the population of spotted owls survive in the long run or if they perish.

## Discrete Time Dynamical Systems (DTDS)
- A Discrete Time Dynamical System (DTDS) or Discrete Dynamical System with $n$ variables, is a sequence $n$-variables $x_k = (x_{1,k},..,x_{n,k})$. Where at each step the next element in the sequence, is a function of the previous term(s) in the sequence.

$$\begin{aligned}x_{1,k} &= f_{1}(x_{1,k-1},...,x_{n,k-1})\\
x_{2,k} &= f_{2}(x_{1,k-1},...,x_{n,k-1})\\
\vdots & \hspace{0.5in} \vdots\\
x_{n,k} &= f_{n}(x_{1,k-1},...,x_{n,k-1})\end{aligned}$$

To simplify the above notation we will just write
$$X_{k} = F(X_{k-1}).$$

A DTDS consists of an updated function(s) $F$ and a initial value $X_0$.

### Examples

- Markov Chains:
    - $X_{k+1} = P X_{k}$ where $P$ is a stochastic Matrix.
- Time Series Forecasting:
    - If we want to predict $x_{t+1}$ using the previous value $x_t$, then $x_{t+1} = f(x_{t})$ is a DTDS.
    - if we have $x_0,...,x_t$ and we want to predict the next entry. Say using the $1$ degree taylor polynomial
    - Let $x'_k = x_{k} - x_{k-1}$, $x''_k = x'_k-x'_{k-1}$
    - Update given by $\displaystyle x_{t+1} = x_t + x'_t(1)^1 + \frac{x''_t}{2}(1)^2$.
- Gradient Descent:
    - Given initial guess $X_0$, we update using $X_{k} = X_{k-1} - h\nabla f(X_{k-1})$ where $h$ is a fixed step size.

### Linear Difference Equations 

A First Order Linear Difference Equation is a special type of Discrete Dynamical System. Given a point $\mathbf{\vec{x}}_k \in \mathbb{R}^n$, the update is given by a matrix $A \in \mathbb{R}^n$ via 
$$\mathbf{\vec{x}}_{k+1} = A \mathbf{\vec{x}}_k$$

<b><u>Note:</u></b> If the matrix $A$ is a stochastic matrix, this is just a Markov Chain.

A **solution** is an explicit expression for $ \vec{x}_k $ rather than a recursive one.

Eigenvalues and eigenvectors are the key to the **long-term behavior** of a dynamical system described by a difference equation. If $\vec{v}_i$ are eigenvectors of $ A $, such that $A \vec{v}_i = \lambda_i \vec{v}_i$ then:
- $\lambda_i^k \vec{v}_i \text{ is a solution}$: each eigenpair gives a different solution.
- $\alpha \lambda_i^k \vec{v}_i + \beta \lambda_j^k \vec{v}_j \text{ is a solution}$: shows linearity of the difference equation
    - If $\vec{x}_k^{(i)} = \lambda_i^k \vec{v}_i$ and $\vec{x}_k^{(j)} = \lambda_j^k \vec{v}_j$ are individually solutions, then their linear combination is also a solution
- $\sum_i a_i \lambda_i^k \vec{v}_i \text{ is a solution}$
    - If the eigenvectors are linearly independent and form a basis.

### Linear Difference Equations 

The **general solution** $\sum_i a_i \lambda_i^k \vec{v}_i$ works when the stage-matrix $A$ is diagonalizable and its eigenvectors form a basis of $\mathrm{R}^n$.

But even if the stage matrix is not diagonalizable, the larger eigenvalues still dominate the long-term behaviour.

### Predator Prey Example

- Wood rats provide up to $80\%$ of the diet for the spotted owl, the main predator of the wood rat.

- Let
$
\begin{pmatrix}
o_k \\
r_k
\end{pmatrix}
$
denote the populations of owls and rats at month $k$. $r_k$ is the number of rats (measured in thousands).

- Furthermore, let 
$$
o_{k+1} = \underbrace{0.5o_k}_{\text{procreation}} + \underbrace{0.4r_k}_{\text{nutrition}}
$$

$$
r_{k+1} = \underbrace{-0.104o_k}_{\text{predation}} + \underbrace{1.1r_k}_{\text{procreation}}
$$

The system can be written in matrix form as

$$
\begin{pmatrix}
o_{k+1} \\
r_{k+1}
\end{pmatrix}
=
\begin{pmatrix}
0.5 & 0.4 \\
-0.104 & 1.1
\end{pmatrix}
\begin{pmatrix}
o_k \\
r_k
\end{pmatrix}.
$$

What would happen in the long run?

### Predator Prey Example
A solution is given by $\mathbf{x}_k = A^k \mathbf{x}_0.$ To understand the long-run behavior, we study the eigenvalues of $A$.

Turns out the two eigenvalues are $\lambda_1=1.02$ and $\lambda_2=0.58$. Corresponding eigenvectors are 
$
\mathbf{v}_1 =
\begin{pmatrix}
10 \\
13
\end{pmatrix}$,
 and 
$
\mathbf{v}_2 =
\begin{pmatrix}
5 \\
1
\end{pmatrix}.
$

An initial vector $\mathbf{x}_0$ can be written as $\mathbf{x}_0 = c_1 \mathbf{v}_1 + c_2 \mathbf{v}_2.$

Then, for $k \ge 0$, 
$
\mathbf{x}_k
= c_1 (1.02)^k \mathbf{v}_1 + c_2 (0.58)^k \mathbf{v}_2
= c_1 (1.02)^k
\begin{pmatrix}
10 \\
13
\end{pmatrix}
+
c_2 (0.58)^k
\begin{pmatrix}
5 \\
1
\end{pmatrix}.
$

As $k \rightarrow \infty$, 
$\mathbf{x}_k \approx c_1 (1.02)^k
\begin{pmatrix}
10 \\
13
\end{pmatrix}$ and the smaller eigenmode disappears asymptotically.

Or, $\mathbf{x}_{k+1} \approx 1.02\mathbf{x}_k$.

### Predator Prey Example

The previous approximation says that eventually both the numbers of owls and rats grow by a factor of almost $1.02$ each month, a $2\%$ monthly growth rate.

Also, $\mathbf{x}_k$ is approximately a multiple of $\begin{pmatrix}
10 \\
13
\end{pmatrix}$. And the entries give us the ratio of the number of owls to rats, i.e., for every $10$ owls there are about $13$ thousand rats.

Generally, if 
$
|\lambda_1| \ge 1 > |\lambda_2| \ge |\lambda_3| \ge \cdots,
$
then
$$\vec{x}_{k+1} \approx \lambda_1 \vec{x}_k,$$

$$
\vec{x}_k \approx c_1 \lambda_1^k \vec{v}_1
$$

Here,
- $ \lambda_1 $ determines the **rate of growth**,  
- $ \vec{v}_1 $ determines the **population ratio**.

## Graphical Description of Solutions
When $A$ is $2 \times 2$, algebraic calculations can be supplemented by a geometric description of a system’s evolution.

We can view the equation 
$
\mathbf{x}_{k+1} = A \mathbf{x}_k
$ 
as a description of what happens to an initial point $\mathbf{x}_0$ in $\mathbb{R}^2$ as it is transformed repeatedly by the system 
$
\mathbf{x} \mapsto A\mathbf{x}.
$

The graph of $\mathbf{x}_0, \mathbf{x}_1, \ldots$ is called a **trajectory** of the dynamical system. Specifically, we care about $3$ main conditions here.

### Origin as an attractor

Plot several trajectories (for different initial conditions) for
$
   \vec{x}_{k+1} = A \vec{x}_k,
   \qquad
   A =
   \begin{pmatrix}
   0.80 & 0 \\
   0 & 0.64
   \end{pmatrix}.
$

   
The eigenvalues and eigenvectors are
$$\lambda_1 = 0.8,
\qquad
\vec{v}_1 =
\begin{pmatrix}
1 \\
0
\end{pmatrix}; 
\lambda_2 = 0.64,
\qquad
\vec{v}_2 =
\begin{pmatrix}
0 \\
1
\end{pmatrix}.
$$

Thus the solution has the form 
$\vec{x}_k = c_1 (0.8)^k\begin{pmatrix}
1 \\
0
\end{pmatrix}
+
c_2 (0.64)^k
\begin{pmatrix}
0 \\
1
\end{pmatrix}.
$

As $k \to \infty$, 
$
\vec{x}_k \longrightarrow
\begin{pmatrix}
0 \\
0
\end{pmatrix}.
$

Here, the origin is an attractor: all trajectories tend toward 
$
\begin{pmatrix}
0 \\
0
\end{pmatrix}
$

because $1 > |\lambda_1| > |\lambda_2|.$

### Origin as a repeller
Plot several trajectories (for different initial conditions) for 
$
\vec{x}_{k+1} = A \vec{x}_k,
\qquad
A =
\begin{pmatrix}
1.44 & 0 \\
0 & 1.20
\end{pmatrix}.
$

The eigenvalues and eigenvectors are 
$$
\lambda_1 = 1.44,
\vec{v}_1 =
\begin{pmatrix}
1 \\
0
\end{pmatrix}; 
\lambda_2 = 1.20,
\vec{v}_2 =
\begin{pmatrix}
0 \\
1
\end{pmatrix}.
$$

Thus the solution has the form 
$\vec{x}_k = c_1 (1.44)^k
\begin{pmatrix}
1 \\
0
\end{pmatrix}+ c_2 (1.20)^k\begin{pmatrix}
0 \\
1
\end{pmatrix}
$.

As $k \to \infty$, 
$
\vec{x}_k \longrightarrow
\begin{pmatrix}
\pm \infty \\
\pm \infty
\end{pmatrix}.
$

Here, the origin is a repeller: all trajectories (except 
$
\begin{pmatrix}
0 \\
0
\end{pmatrix}
$) are unbounded and flow away from the origin because $|\lambda_1| > |\lambda_2| > 1.$

### Origin as a saddle point
Plot several trajectories (for different initial conditions) for 
$
\vec{x}_{k+1} = A \vec{x}_k,
\qquad
A =
\begin{pmatrix}
2.0 & 0 \\
0 & 0.5
\end{pmatrix}.
$

The eigenvalues and eigenvectors are
$$
\lambda_1 = 2.0,
\qquad
\vec{v}_1 =
\begin{pmatrix}
1 \\
0
\end{pmatrix}, 
\lambda_2 = 0.5,
\qquad
\vec{v}_2 =
\begin{pmatrix}
0 \\
1
\end{pmatrix}.
$$

Thus the solution has the form 
$ \vec{x}_k= c_1 (2.0)^k\begin{pmatrix}
1 \\
0
\end{pmatrix} + c_2 (0.5)^k\begin{pmatrix}
0 \\
1
\end{pmatrix}.
$

As $k \to \infty$, 
$
\vec{x}_k \longrightarrow
\begin{cases}
\begin{pmatrix}
\pm \infty \\
0
\end{pmatrix} & \text{if } c_1 \neq 0, \\
\\
\begin{pmatrix}
0 \\
0
\end{pmatrix} & \text{if } c_1 = 0.
\end{cases}
$

Here, the origin is a saddle point: trajectories that start with an $x_1 \neq 0$ component are unbounded (and flow away from the origin) along the $x_1$-axis, while $x_2 \to 0$ because $|\lambda_1| > 1 > |\lambda_2|$.

## Complex Eigenvalues 

When a real $2\times2$ matrix $A$ has complex eigenvalues, $A$ is not diagonalizable on $\mathcal{R}^2$ but the dynamical system $\mathbf{x}_{k+1} = A\mathbf{x}_{k}$ can still be described.

$\color{blue}{\textbf{Complex Numbers}}$: Recall that complex numbers are of the form $\mathbf{z}= a+ib$ with two parts
- $a$ is the **real** part
- $b$ is a real constant, $i^2=-1$ and $ib$ is the **imaginary** part

You can add and multiply complex numbers.

You can also take the absolute value of an imaginary number–consider it to be the distance from $0$, $abs(\mathbf{z}) = \lvert a + ib \rvert = \sqrt{a^2+b^2}$

### Complex Eigenvalues 

Rotation is associated with complex eigenvalues.
- If $A$ has two complex eigenvalues whose absolute value is less than $1$:

  - The origin is an **attractor**.
  - $\vec{x}_k$ spirals around the origin along an elliptical trajectory.

- If $A$ has two complex eigenvalues whose absolute value is greater than $1$:

  - The origin is a **repeller**.
  - $\vec{x}_k$ spirals away from the origin along an elliptical trajectory.

### Complex Eigenvalues

Consider the matrix 
$
A =
\begin{pmatrix}
0.8 & 0.5 \\
-0.1 & 1.0
\end{pmatrix}
$.

It has eigenvalues $0.9 \pm 0.2i$, with eigenvectors $\begin{pmatrix}
1 \mp 2i \\
1
\end{pmatrix}
$. 
We can trace three trajectories of the system $\mathbf{x}_{k+1} = A \mathbf{x}_k$, with $3$ different initial vectors 
$\begin{pmatrix}
0 \\
2.5
\end{pmatrix},  
\begin{pmatrix}
3 \\
0
\end{pmatrix},  
\begin{pmatrix}
0 \\
-2.5
\end{pmatrix}.
$

## So what happened to the Spotted Owls?

Recall from earlier, that the population dynamics were represented by:
- a vector $\mathbf{x}_k = (j_k, s_k, a_k)$, where $j_k$, $s_k$, and $a_k$ are the numbers of females in the juvenile, subadult, and adult stages
- a stage-matrix 
$A = \begin{pmatrix}
0 & 0 & 0.33 \\
0.18 & 0 & 0 \\
0 & 0.71 & 0.94
\end{pmatrix}$

To find the long-term behaviour of the population we find the eigenvalues and eigenvectors of $A$. Here we find three eigenvalues.
$$
\lambda_1 = 0.98; \lambda_2=  -.02 + 0.21i; \lambda_2=  -.02 - 0.21i
$$
All 3 eigenvalues have magnitude less than 1. Specifically the magnitude of the complex eigenvalues is: 
$$\lvert \lambda_2 \rvert^2 = \lvert \lambda_3 \rvert^2 = (-.02)^2 + (0.21)^2 = 0.445$$
We can plot the growth rates and the ratios of juveniles and subadults to adults as a function of time. Unfortunately since the largest real eigenvalue $\lvert \lambda_1 \rvert < 1$, this model implies that the **spotted owls will perish** &#128577; !

### Is there any hope for the Spotted Owls?

In the previous stage-matrix the $18\%$ entry comes from the fact that although $60\%$ of the juvenile owls live long enough to leave the nest and search for new home territories, only $30\%$ of that group *survive the search* and find new home ranges as sub-adults.

If the search survival rate of the juvenile owls *increases to $50\%$* then $A(2,1)$ increases from $0.18$ to $0.3$. 

Let's now reevaluate the predictions of the updated stage-matrix for the spotted own population. Now the eigenvalues of $A$ turn out to be approximately,
$$
\lambda_1 = 1.01; \lambda_2=  -.03 + 0.26i; \lambda_2=  -.03 - 0.26i
$$
Assuming the respective eigenvectors are $\mathbf{v_1}, \mathbf{v_2}, \mathbf{v_3}$, we can rewrite the solution of the dynamical system as:
$$
\mathbf{x_k} = c_1(1.01)^k\mathbf{v_1} +c_2(-.03 + 0.26i)^k\mathbf{v_2} +c_3(-.03 - 0.26i)^k\mathbf{v_3}
$$

### Is there any hope for the Spotted Owls?

As $k \rightarrow \infty$, the last two *complex* vectors from the right tend to zero and $x_k$ resembles the *real* vector $c_1(1.01)^k\mathbf{v_1}$.

The eigenvector $\vec{v_1} = \begin{pmatrix}
10\\
3\\
31
\end{pmatrix}$ describes the eventual distribution of the owls by life-stages:  for every $31$ adults, there will be about $10$ juveniles and $3$ subadults. 

Thus the owl population will grow slowly, with a long-term growth rate of $1.01$ meaning the population survive &#128512; !
